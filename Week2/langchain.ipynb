# Simple usage of langchain
Let's translate a text from English to German

```python
from IPython.display import Markdown, display
from dotenv import load_dotenv, find_dotenv
from langchain.chat_models import init_chat_model
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

load_dotenv(find_dotenv(), override=True)

model = init_chat_model("gemini-3-flash-preview", model_provider="google_genai")
messages = [
    SystemMessage("Translate the following from English into German"),
    HumanMessage("Hello Germany!")
]
response: AIMessage = model.invoke(messages)
display(Markdown(response.text))
```

Now, let's use a variable to achieve the same thing above.

```python
from langchain_core.prompts import ChatPromptTemplate

system_template = "Translate the following from English into {language}"

prompt_template = ChatPromptTemplate.from_messages(
    [SystemMessage(system_template), HumanMessage("{text}")]
)
```

```python
prompt = prompt_template.invoke({ "language": "German", "text": "Hello Germany!" })
prompt
```

```python
response = model.invoke(prompt)
display(Markdown(response.text))
```
